# Concatenating 3 datasets into 1

```python
import pandas as pd
```

**Dataset No. 1**

```python
data_1 = pd.read_csv("/content/train.csv")
data_1
```

Label meaning:
*   0 -> normal
*   1 -> hateful

```python
data_1_updated = data_1.drop(["id"], axis=1)
data_1_updated 
```

**Dataset No. 2**

```python
data_2 = pd.read_csv("/content/hate-speech-and-offensive-language.csv")
data_2
```

Label meaning:
*   0 -> normal
*   1 -> hateful
*   2-> offensive


---


Previously the order was *0->hateful, 1->offensive, 2->neither*, but it was changed so that it is compatible with the other tables

```python
data_2_updated = data_2.drop(["Unnamed: 0", "count", "hate_speech",	"offensive_language",	"neither"], axis=1)
data_2_updated = data_2_updated.rename(columns={"class": "label"})
data_2_updated["label"] = data_2_updated["label"].replace("0", "1")
data_2_updated["label"] = data_2_updated["label"].replace("1", "2")
data_2_updated["label"] = data_2_updated["label"].replace("2", "0")
data_2_updated
```

**Dataset No. 3**

```python
data_3 = pd.read_csv("/content/en_dataset_with_stop_words.csv")
data_3
```

Label meaning:

*   0 -> normal
*   1 -> hateful
*   2-> offensive

```python
data_3_updated = data_3.drop(["HITId", "directness",	"annotator_sentiment",	"target",	"group"], axis=1)
data_3_updated = data_3_updated.rename(columns={"sentiment": "label"})
data_3_updated = data_3_updated.reindex(columns=["label", "tweet"])
data_3_updated
```

All the in between tweets were deleted.

```python
value_counts = data_3_updated["label"].value_counts()
value_counts
```

```python
selected_rows = data_3_updated["label"].isin(["offensive", "hateful", "normal"])
data_3_updated = data_3_updated.drop(index=data_3_updated[~selected_rows].index)
data_3_updated["label"] = data_3_updated["label"].replace("normal", "0")
data_3_updated["label"] = data_3_updated["label"].replace("hateful", "1")
data_3_updated["label"] = data_3_updated["label"].replace("offensive", "2")
data_3_updated
```

```python
data_1_updated=data_1_updated.to_csv("data_1.csv", index=False)
data_2_updated=data_2_updated.to_csv("data_2.csv", index=False)
data_3_updated=data_3_updated.to_csv("data_3.csv", index=False)
```

```python
table1 = pd.read_csv("data_1.csv")
table2 = pd.read_csv("data_2.csv")
table3 = pd.read_csv("data_3.csv")

merged_table = pd.concat([table1, table2, table3], ignore_index=True)

merged_table.to_csv("merged_table.csv", index=False)
```
